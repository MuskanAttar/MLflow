```python
import pandas as pd
from sklearn import datasets
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
import mlflow
from mlflow.models import infer_signature
```

```python
## load the dataset
X,y=datasets.load_iris(return_X_y=True)
# split the data into training and test sets
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20)

# Define the model hyperparameters
params = {"penalty":"l2","solver": "lbfgs", "max_iter": 1000, "multi_class": "auto", "random_state": 8888}

##train the model

lr=LogisticRegression(**params)
lr.fit(X_train,y_train)
```

```python
## Prediction on the test set
y_pred=lr.predict(X_test)
y_pred
```

```python
accuracy=accuracy_score(y_test,y_pred)
print(accuracy)
```

```python
### MLFLOW tracking
mlflow.set_tracking_uri(uri="http://127.0.0.1:5000")

##create a new MLFLOW experiment
mlflow.set_experiment("MLFLOW Quickstart")

## Sstart an MLFLOW run
with mlflow.start_run():
    ## log the hyperparameters
    mlflow.log_params(params)

    ## Log the accuracy metrics
    mlflow.log_metric("accuracy",accuracy)

    # Set a tag that we can use to remind ourselves what this run was for
    mlflow.set_tag("Training Info", "Basic LR model for iris data")

    ## Infer the model signature

    signature=infer_signature(X_train,lr.predict(X_train))

    ## log the model
    model_info=mlflow.sklearn.log_model(
        sk_model=lr,
        artifact_path="iris_model",
        signature=signature,
        input_example=X_train,
        registered_model_name="tracking-quickstart",

    )


    
```

# infer_signature :
To record the structure (schema) of the model's inputs and outputs for logging and future reference, especially for deployment.

üîß How it works:
X_train: the input data used to train the model (usually a DataFrame or array).

lr.predict(X_train): the output predictions from the model using that same input.

üì¶ What infer_signature does:
It returns a model signature, which is an object that defines:

The types and shapes of the input features (X_train)

The types and shapes of the outputs (lr.predict(X_train))

This signature is later saved when logging the model, ensuring that:

You or someone else can re-use or serve the model correctly.

MLflow knows what shape and data types to expect when making predictions.
